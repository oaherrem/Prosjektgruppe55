```python
import json
import sys
import time
from datetime import datetime, timedelta

sys.path.insert(0, '../src/')

import metar_writer as mw

# Parametere for METAR
icao_codes = ["ENBO", "ENOL"]
start_date = datetime.strptime("2024-01-01", "%Y-%m-%d-")
end_date = datetime.strptime("2024-12-31", "%Y-%m-%d")

# Generer en liste over datoer
dates = [(start_date + timedelta(days=i)).strftime("%Y-%m-%d") for i in range((end_date - start_date).days)]

for icao_code in icao_codes:
    all_data = []  # Liste for å samle alle dataene for én ICAO-kode

    for date in dates:
        try:
            # Henter informasjon for gitte parametere
            metar = mw.get_metar(icao_code, date)

            # Splitt METAR-dataene ved linjeskift og fjerner eventuelle tomme linjer
            metar_lines = [line for line in metar.split('\n') if line.strip ()]

            # Samler dataene i en JSON-struktur
            data_metar = {
                "metar": metar_lines
            }
            all_data.append(data_metar)

            # Legg inn en forsinkelse for å unngå overbelastning av API-en
            time.sleep(1)  # 1 sekund forsinkelse mellom forespørsler

        except Exception as e:
            print(f"Feil ved henting av data for {icao_code} på {date}: {e}")

    # Lagrer alle dataene for denne ICAO-koden i én JSON-fil
    output_file = f"../data/json/{icao_code}_metar_data.json"
    with open(output_file, "w") as json_file:
        json.dump(all_data, json_file, indent=4)

    print(f"Alle dataene for {icao_code} er lagret i {output_file}")
```

#### Datainnsamling

- Programmet vi har laget henter funksjoner fra mappen 'src', setter parameterene for data fra en åpen API-kilde og lagrer dataen som en JSON-fil under mappen 'data'. Den henter ut fra en bestemt start-dato til en bestemt slutt-dato. For vår analyse har vi hentet et år med data for hver flyplass, som skal gi oss et godt grunnlag til å se på værdata, spesielt vindretning og vindhastighet gjennom året. Vi har blitt nødt til å legge inn en delay i for-loopen slik at vi ikke overbelaster API.met, da dette er en forespørsel av data på over 35 tusen linjer.

- Vi har valgt en sentral og relevant åpen datakilde for miljødata; api.met.no. Tjenesten leveres av Meterologisk institutt, og er offisiell og pålitelig kilde for meterologiske data fra Norge. 

- Vi har valgt å spesifisere oppgaven inn på Aviation Weather, mer spesifikt Meteorological Aerodrome Report (METAR). METAR blir publisert hver halvtime og er en vær observasjon. METAR er et internasjonal standardisert format, som brukes i luftfarten. Datane er tilgjengelig og krever ingen API-nøkkel som gjør data-innsamlingen enkelt. Parameterene for API-en er tid, lokasjon og formatvalg, som vi ser som en fordel for vår analyse.
