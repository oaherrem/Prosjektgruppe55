#### Beregner statistiske mål som gjennomsnitt, median og standardavvik
Vi benytter oss av Numpy og Panda's innebygde koder for å regne ut gjennomsnitt, median og standardavvik for data som vi henter ut fra datasettet vårt. Da vi har valgt å fokusere på vind-retning og hastighet er disse verdiene viktig statistiske mål i denne analysen, fordi dette forteller oss senere hvor mye målingene varierer. 


```python
# Gjennomsnitt av interessant data 
avg_wind_direction = df_airport["Wind_direction"].mean().round(1)
avg_wind_speed = df_airport["Wind_speed"].mean().round(1)
avg_gust_speed = df_airport["Gust_speed"].mean().round(1)
avg_QNH = df_airport["QNH"].mean().round(1)

# Median
median_wind_direction = df_airport["Wind_direction"].median().round(1)
median_wind_speed = df_airport["Wind_speed"].median().round(1)
median_gust_speed = df_airport["Gust_speed"].median().round(1)
median_QNH = df_airport["QNH"].median().round(1)

# Standardavvik
std_wind_direction = df_airport["Wind_direction"].std().round(1)
std_wind_speed = df_airport["Wind_speed"].std().round(1)
std_gust_speed = df_airport["Gust_speed"].std().round(1)
std_QNH = df_airport["QNH"].std().round(1)
```

Utskrift av statistiske mål

```python
print(f"Beregner gjennomsnitt av interessant data for {airport}.")
print("Gjennomsnittlig vindretning:", avg_wind_direction)
print("Gjennomsnittlig vindhastighet:", avg_wind_speed)
print("Gjennomsnittlig vindkasthastighet:", avg_gust_speed)
print("Gjennomsnittlig trykk:",avg_QNH)
print("Median vindretning:", median_wind_direction)
print("Median vindhastighet:", median_wind_speed)
print("Median vindkasthastighet:", median_gust_speed)
print("Median trykk:", median_QNH)
print("Standardavvik vindretning:", std_wind_direction)
print("Standardavvik vindhastighet:", std_wind_speed)
print("Standardavvik vindkasthastighet:", std_gust_speed)
print("Standardavvik trykk:", std_QNH)
```

Det er et enormt datasett og denne framstillingen av statistiske mål gir oss en bredere forståelse av dataen vi jobber med. Dette er høyst nødvendig for å lykkes med de neste punktene som visualisering og prediktiv analyse.

Vi ser for eksempel at standardavviket på vinddata er ganske høyt noe som gjør det utfordrende å prediktere, likevel gir median og gjennomsnittlig-vinddata oss en god indikator på hvilken retning vinden som regel kommer fra. Dette ser vi nærmere på i visualiseringen i neste del.

### Sidevind analyse
En annen analyse vi er interessert i er hvor mange tilfeller av sidevind som oppstår på en flyplass over en periode. Grunnen til dette er fordi det påvirker flyplassen sin operasjonelle status. Vi har valgt å lage en dynamisk funskjon "is_crosswind", som sjekker alle vindretningene som ligger innenfor vårt definerte kritiske området, samt vindhastighet over 35 knop. 

Henter inn funksjonen "is_crosswind" fra src mappa og analyserer hvor mange tilfeller av datapunktene våre som har en vindretning og vindhastighet som medfører sidevind. 

```python
#henter fil fra src mappa.
import sys
sys.path.insert(0, '../src/')

from crosswind import is_crosswind
```

```python
runway = 330 # beslutningsvariabel - rullebanevinkel til ENOL, kan bytte ut denne med 150 og får like mange tilfeller. 
wind_data = df_airport # data fra csv-fil 
crosswind_rows = [] # for å kunne visualisere det i en pandas.core.frame.Dateframe 
count = 0
# for-løkke som itererer over hver rad i datasettet. 

for _, datapoint in wind_data.iterrows():
    if is_crosswind(runway, datapoint["Wind_direction"], datapoint["Wind_speed"]):
        count += 1 # summerer antall tilfeller 
        crosswind_rows.append(datapoint)

# printer ut antall tilfeller med crosswind
print(f"Printer det totale antallet av tilfeller med crosswind som er lik: {count}")

# lager en data frame av crosswind_rows
crosswind_df = pd.DataFrame(crosswind_rows)
crosswind_df
```

#### Optimal rullebaneretning i forhold til sidevind - bruk av historisk data

```python
runway = 330 # beslutningsvariabel - rullebanevinkel til ENOL, kan bytte ut denne med 150 og få like mange tilfeller. 
wind_data = df_airport #data fra csv-fil 
best_crosswind_rows = [] #for å kunne visualisere det i en pandas.core.frame.Dateframe 
best_count = float('inf')
optimal_runway = -1

#for-løkke som itererer over hver rad i datasettet. 
for angles in range(0, 180):
    count = 0
    temp_crosswind_rows = []

    for _, datapoint in wind_data.iterrows():
        if is_crosswind(angles, datapoint["Wind_direction"], datapoint["Wind_speed"]):
            count += 1 # summerer antall tilfeller 
            temp_crosswind_rows.append(datapoint)

    if count < best_count:
        best_count = count 
        optimal_runway = angles 
        best_crosswind_rows = temp_crosswind_rows

#print(f"Printer det totale antallet av tilfeller med sidevind som er lik: {count}")
print(f"Ved rullebaneretning: {optimal_runway}, er det {best_count} tilfeller av sidevind ved {airport}.")
crosswind_df = pd.DataFrame(crosswind_rows)

crosswind_df.head(5)

```
