I del 1 av Oppgave 3 ser vi en data frame som henter ut et sample av 10 linjer med data og plasserer de i sine respektive kolonner. Det har vi løst med list comprehension. Vi ser også at manglende verdier blir fylt med NaN. 

I vårt prosjekt ser vi i hovedsak på vind-retning og -hastighet. Dette er verdier **. Derfor kjører vi en annen funksjon nedenfor som eksluderer irrelevante kolonner for vårt prosjekt.

```python
import sys, os
import pandas as pd

sys.path.insert(0, '../src/')
#henter dataFrame_metar.py fil fra src mappa.
import dataFrame_wind as dataframe

#variabler 
column_names = ["Airport", "Date/time", "Wind_direction", "Wind_speed", "Gust_speed", "QNH"]

#Henter filene fra data mappa
file_paths = {
    "ENOL": '../data/json/ENOL_metar_data.json',
    "ENBO": '../data/json/ENBO_metar_data.json'
}

dfs = {}

#Itererer gjennom filene i file_paths
for airport, file_path in file_paths.items():
    if not os.path.exists(file_path):
        print(f"Filen {file_path} finnes ikke")
    else:
        df_m = pd.read_json(file_path)
        dfs[airport] = dataframe.metar_df(column_names, df_m)
        print(f"filene {file_path} er lagret i dataframe for {airport}")

df_ENBO = dfs.get("ENBO")
df_ENOL = dfs.get("ENOL")
df_ENOL.to_csv('../data/csv/ENOL_metar_data.csv', index = False)
df_ENBO.to_csv('../data/csv/ENBO_metar_data.csv', index = False)
print("data frame av ENOL som inneholder Airport, Date/time, Wind_direction, Wind_speed, Gust_speed, er lagert i folder: data" )
print("data frame av ENBO som inneholder Airport, Date/time, Wind_direction, Wind_speed, Gust_speed, lagert i folder: data" )
```
